```python
import pandas
import numpy as np
```

```python
r_cols = ["user_id","movie_id","rating"]
ratings = pandas.read_csv("datasets/movie_rating/u.data",sep="\t",names=r_cols,usecols=range(3))
```

```python
ratings
```

```python
m_cols = ["movie_id","title"]
movies = pandas.read_csv("datasets/movie_rating/movie.data",sep="|",names=m_cols,usecols=range(2))
```

```python
movies
```

```python
movie_rating = pandas.merge(ratings, movies)
movie_rating
```

```python
userRatings = movie_rating.pivot_table(index=["user_id"],columns=["title"],values='rating')
userRatings
```

```python
corrMatrix = userRatings.corr()
corrMatrix
```

```python
corrMatrix = userRatings.corr(method="pearson",min_periods=100)
corrMatrix.head()
```

```python
myratings = userRatings.loc[0].dropna()
myratings
```

```python
sim_all = []
```

```python
for i in range(0,len(myratings.index)):
    sims = corrMatrix[myratings.index[i]].dropna() * myratings.iloc[i]
    sim_all.append(sims)
```

```python
sim_ss = pandas.Series()
sim_ss = pandas.concat(sim_all)
```

```python
sim_ss.sort_values(inplace=True,ascending=False)
sim_ss
```

```python
sim_ss = sim_ss.groupby(sim_ss.index).sum()
```

```python
sim_ss.sort_values(inplace=True,ascending=False)
sim_ss.head(10)
```

```python
filterSim = sim_ss.drop(myratings.index)
filterSim
```
