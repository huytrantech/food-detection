```python
import pandas
import numpy as np
```

```python
r_cols = ["user_id","movie_id","rating"]
ratings = pandas.read_csv("datasets/movie_rating/u.data",sep="\t",names=r_cols,usecols=range(3))
```

```python
ratings
```

```python
m_cols = ["movie_id","title"]
movies = pandas.read_csv("datasets/movie_rating/movie.data",sep='|',names=m_cols,usecols=range(2))
```

```python
movies
```

```python
merge_table = pandas.merge(movies,ratings)
```

```python
merge_table
```

```python
movie_rating_pivot = merge_table.pivot_table(index=["user_id"],columns=["title"],values='rating')
```

```python
movie_rating_pivot.head()
```

```python
starwar_rating = movie_rating_pivot["Star Wars (1977)"]
```

```python
starwar_rating
```

```python
similar_movies = movie_rating_pivot.corrwith(starwar_rating)
```

```python
similar_movies = similar_movies.dropna()
```

```python
similar_movies
```

```python
similar_movies.sort_values(ascending=False).head(20)
```

```python
movieStats = merge_table.groupby('title').agg({'rating':{np.size,np.mean}})
```

```python
movieStats
```

```python
popularMovies = movieStats['rating']['size'] >= 100
```

```python
movieStats[popularMovies].sort_values([('rating','mean')],ascending=False)[:15]
```

```python
movieStats[popularMovies].sort_values([('rating','size')],ascending=False)
```

```python
df_similarity = pandas.DataFrame(similar_movies,columns=['similarity'])
```

```python
df_similarity
```
